Tokenization ans stopwatch removal

```python
import nltk

nltk.download('punkt')

from nltk.tokenize import sent_tokenize, word_tokenize

# Sample text
text = "It was a very pleasant day, the weather was cool and there were showers. I went to market to buy some fruits."

# Tokenize the text into sentences
sents = sent_tokenize(text)

# Print the tokenized sentences
print(sents)
words_list=word_tokenize(sents[0])
```

```python
print(words_list)
```

stopwords removal do not take part much in sentence
 a,the,ans,it
 not making much weight in the sentence

```python
import nltk
from nltk.corpus import stopwords
#nltk.download('stopwords')
sw = set(stopwords.words('english'))
print(sw)
print(len(sw))
```

remove stopwords from sentence

```python
def filter_words(words_list):
  useful_words=[w for w in words_list if w not in sw]
  return useful_words
```

```python
useful_words=filter_words(words_list)
print(useful_words)
```

regular expression tokenization to remove . ,

```python
from nltk.tokenize import RegexpTokenizer
tokenizer=RegexpTokenizer("[, .]")
sents="send the 50 documents to abc,def,ghi,jkl"
print(tokenizer.tokenize(sents))
#donot want , and .
tokenizer=RegexpTokenizer("[a-z0-9]+")
print(tokenizer.tokenize(sents))
```
